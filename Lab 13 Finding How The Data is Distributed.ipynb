### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
## Write your code here

print(df.columns)

print(df.describe())
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here

missing_values = df.isnull()
print("Missing values:\n", missing_values)

missing_counts = df.isnull().sum()
print("\nMissing value counts per column:\n", missing_counts)

mode_RemoteWork = df['RemoteWork'].mode()[0]
df['RemoteWork'].fillna(mode_RemoteWork, inplace = True)

print(df['RemoteWork'].isnull().sum())

mean_JobSat = df['JobSat'].mean()
df['JobSat'].fillna(mean_JobSat, inplace = True)

print(df['JobSat'].isnull().sum())

print(df['JobSat'])
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
## Write your code here

value_counts_Employment = df['Employment'].value_counts()
print(value_counts_Employment)

value_counts_JobSat = df['JobSat'].value_counts()
print(value_counts_JobSat)

value_counts_YearsCodePro = df['YearsCodePro'].value_counts()
print(value_counts_YearsCodePro)

print(df['YearsCodePro'].nunique())
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here

# Create value count for JobSat
counts = df['JobSat'].value_counts()

# Plot the Pie Chart
plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
plt.title('Distribution of Job Satisfaction')
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Explode the language lists into individual rows
df_exploded = df.explode('LanguageHaveWorkedWith').explode('LanguageWantToWorkWith')

# Count occurrences separately
worked_counts = df_exploded['LanguageHaveWorkedWith'].value_counts()
want_counts = df_exploded['LanguageWantToWorkWith'].value_counts()

# Merge the two counts into a single DataFrame
lang_comparison = pd.DataFrame({'WorkedWith': worked_counts, 'WantToWorkWith': want_counts}).fillna(0)

# Select top languages
top_n = 15  # Adjust as needed
lang_comparison = lang_comparison.nlargest(top_n, 'WorkedWith')

# Plot grouped bar chart
lang_comparison.plot(kind='bar', figsize=(12, 6), width=0.8)

# Customize plot
plt.title('Top Programming Languages: Worked With vs Want to Work With')
plt.xlabel('Programming Language')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.legend(title="Category")

plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here



# Convert YearsCodePro to numeric (handling non-numeric values)
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Remove NaN values
df_filtered = df[['YearsCodePro', 'JobSat']].dropna()


pearson_corr = df_filtered['YearsCodePro'], df_filtered['JobSat']
print(pearson_corr)
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here

# Create cross-tabulation between Employment and Education Level
employment_education_crosstab = pd.crosstab(df['Employment'], df['EdLevel'])

# Increase figure size and rotation for better readability
plt.figure(figsize=(15, 8))  # Larger figure size
ax = employment_education_crosstab.plot(kind='bar', stacked=True, figsize=(15, 8))

# Customize plot
plt.title('Employment vs Education Level', fontsize=14)
plt.xlabel('Employment Status', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Rotate x-axis labels at a readable angle
plt.xticks(rotation=30, ha='right', fontsize=10)  

# Adjust legend placement
plt.legend(title='Education Level', bbox_to_anchor=(1.05, 1), loc='upper left')

# Add spacing for better layout
plt.tight_layout()

# Show plot
plt.show()
```
